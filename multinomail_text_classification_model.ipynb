```python
x = ["This was an awesome movie",
     "Great movie! I liked it a lot",
     "Happy Ending! awesome acting by the hero",
     "loved it! truly great",
     "bad not upto the mark",
     "could have better",
     "Surely a Disappointing movie"]

y = [1,1,1,1,0,0,0] # 1 - Positive, 0 - Negative Class
```

```python
x_test = ["I was happy & happy and I loved the acting in the movie",
          "The movie I saw was not bad"]
```

```python
### cleaning
import clean_text as ct
```

```python
x_clean=[ct.get_stemmed_review(i) for i in x]
xt_clean=[ct.get_stemmed_review(i) for i in x_test]
```

```python
print(x_clean)
```

# Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(ngram_range=(1,2))# for case where we need to consider bad as well
```

```python
x_vec = cv.fit_transform(x_clean).toarray()
print(cv.get_feature_names_out())
print(x_vec)
print(x_vec.shape)
```

```python
# we only trainsform on test data and not fit as fit is for the train data so test vector has same columns
```

```python
xt_vec=cv.transform(xt_clean).toarray()
print(xt_vec)
cv.get_feature_names_out()
```

```python
#multinomial naive bayes
```

```python
from sklearn.naive_bayes import MultinomialNB,BernoulliNB
```

```python
mb= MultinomialNB()
```

```python
# train
mb.fit(x_vec,y)
```

```python
# predict
mb.predict(xt_vec)
```

```python
mb.predict_proba(xt_vec) # prob for each reviw class 
```

```python
bnb=BernoulliNB(binarize=0.0)#alpha for smoothing binarize is by default 0.0 for changin count >=1 to 1 and coutn=0 to 0
```

```python
bnb.fit(x_vec,y)
```

```python
bnb.predict(xt_vec)
```

```python
bnb.predict_proba(xt_vec)
```

```python
bnb.score(x_vec,y)
```
